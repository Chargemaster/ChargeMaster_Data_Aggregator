```python
import pandas as pd
```

Garfield County Hospital Disctrict

```python
gar = pd.read_csv("files/Garfield_County_Hospital_District_4.csv", header=2, usecols=[0,1,8])
gar.columns = ['drg_code', 'name', 'price']
gar.insert(0, 'hospital', 'Arbor Health')
gar.insert(1, 'county', 'Garfield')
gar.insert(2, 'hospital_size', 'Medium')
gar
```

Arbor Health

```python
arb = pd.read_csv("files/Arbor_Health_2.csv", header=2, usecols=[0,1,8])
arb.columns = ['drg_code', 'name', 'price']
arb.insert(0, 'hospital', 'Arbor Health')
arb.insert(1, 'county', 'Lewis')
arb.insert(2, 'hospital_size', 'Medium')
arb
```

Forks Community Hospital

```python
fork = pd.read_csv("files/Forks_Community_Hospital_3.csv", header=2, usecols=[0,1,8])
fork.columns = ['drg_code', 'name', 'price']
fork.insert(0, 'hospital', 'Forks Community Hospital')
fork.insert(1, 'county', 'Clallam')
fork.insert(2, 'hospital_size', 'Medium')
fork
```

East Adams Rural Healthcase

```python
east = pd.read_csv("files/East_Adams_Rural_Healthcare_2.csv", header=2, usecols=[0,1,8])
east.columns = ['drg_code', 'name', 'price']
east.insert(0, 'hospital', 'East Adams Rural Healthcare')
east.insert(1, 'county', 'Adams')
east.insert(2, 'hospital_size', 'Medium')
east
```

Fairfax Behavioral Health Everett

```python
fairfax = pd.read_excel("files/Fairfax_Behavioral_Health_Everett_4.xlsx", header=4)
fairfax.columns = ['drg_code', 'name', 'price']
fairfax.insert(0, 'hospital', 'Fairfax Behaviroal Heath Everett')
fairfax.insert(1, 'county', 'Snohomish')
fairfax.insert(2, 'hospital_size', '')
fairfax
```

Jefferson Healthcare

```python
jeff = pd.read_excel("files/Jefferson_Healthcare_2.xlsx", header=3, usecols=[3,4,9])
jeff.columns = ['drg_code', 'name', 'price']
jeff.insert(0, 'hospital', 'Jefferson Healthcare')
jeff.insert(1, 'county', 'Jefferson')
jeff.insert(2, 'hospital_size', 'Medium')
jeff
```

Merge all

```python
hosp = [gar, arb, fork, east, fairfax, jeff]
dt = pd.concat(hosp)
dt
```
